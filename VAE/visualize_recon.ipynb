```python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
recon_dir = "/space/mcdonald-syn01/1/projects/jsawant/Diffusion_paper/VAE_GAN/best_vae_gan/reconstructions"
files = glob.glob("/space/mcdonald-syn01/1/projects/jsawant/Diffusion_paper/VAE_GAN/best_runs/vae_gan_run_20250226_161525/reconstructions/*_original.npy")
example_orig_file = "mwp1sub-UNAM36966_ses-01_T1w_original.npy"
example_recon_file = example_orig_file.replace('original', 'recon')
```

```python
files[0]
```

```python
i = 2
original = np.load(files[i])
recon = np.load(files[i].replace('original', 'recon'))
# original = np.load(os.path.join(recon_dir,example_orig_file)) #
# recon = np.load(os.path.join(recon_dir,example_recon_file)) #
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))


# Show middle slice of the volume
middle_slice = 70 # original.shape[2] // 2
fig.suptitle(f"Coronal View, Slice {middle_slice}", fontsize=16)

# Plot original (rotated left by 90 degrees)
im1 = ax1.imshow(np.rot90(original[0, :, middle_slice, :]), cmap='gray')
ax1.set_title('Original')
plt.colorbar(im1, ax=ax1)

# Plot reconstruction (rotated left by 90 degrees)
im2 = ax2.imshow(np.rot90(recon[0, :, middle_slice, :]), cmap='gray')
ax2.set_title('Reconstruction')
plt.colorbar(im2, ax=ax2)

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to fit title
plt.show()
```
